```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
dummies=pd.get_dummies(df['ST_Slope'])
dummies
```

```python
df=pd.concat([df,dummies],axis='columns')
df.drop(['Up','ST_Slope'],axis=1,inplace=True)
df
```

```python
dummies=pd.get_dummies(df.Sex)
dummies.drop('M',axis=1,inplace=True)
dummies
```

```python
df=pd.concat([df,dummies],axis='columns')
df.drop('Sex',axis=1,inplace=True)
df
```

```python
dummy1=pd.get_dummies(df['ExerciseAngina'])
dummy1.drop('N',axis=1,inplace=True)
df=pd.concat([df,dummy1],axis=1)
df
```

```python
dummy1=pd.get_dummies(df['RestingECG'])
dummy1.drop('Normal',axis=1,inplace=True)

df=pd.concat([df,dummy1],axis=1)
df.drop(['ExerciseAngina','RestingECG'],axis=1,inplace=True)
df
```

```python
dummy1=pd.get_dummies(df['ChestPainType'])
dummy1.drop('TA',axis=1,inplace=True)
df=pd.concat([df,dummy1],axis=1)
df.drop('ChestPainType',axis=1,inplace=True)
df
```

```python
from sklearn.decomposition import PCA
X=df.drop('HeartDisease',axis=1)
y=df.HeartDisease
X
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X.keys())
X_scaled
```

```python
pca=PCA(.9)
```

```python
X_scaled.shape
```

```python
X_scaled=pca.fit_transform(X_scaled)
X_scaled.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
X_trian,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=30)
```

```python
model.fit(X_trian,y_train)
```

```python
model.score(X_test,y_test)
```
